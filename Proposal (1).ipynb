```python
import csv
import re
import functools as ft
import numpy as np

def open_csv(file_name):
    collection = []
    with open(file_name, 'r') as f:
        for line in csv.DictReader(f):
            collection.append(line)
        return collection
    
students_math = open_csv('student-mat.csv')
students_por = open_csv('student-por.csv')
print(students_math[0])
```

```python
# geting to know data in student-mat.csv
num_elements_math = len(students_math)
print("Number of Elements: %d"%num_elements_math)

list_of_ages_math = list(map(lambda x:int(x['age']),students_math))
set_of_ages_math = sorted(list_of_ages_math)
#Range of Ages
print("Lowest Age:   %d"%set_of_ages_math[0])
print("Highest Age: %d"%set_of_ages_math[len(set_of_ages_math)-1])
print("Range:        %d"%(set_of_ages_math[len(set_of_ages_math)-1]-set_of_ages_math[0]))
# average of age
mean_ages_math = ft.reduce(lambda a,b: a+b,list_of_ages_math)/len(list_of_ages_math)
print("\nMean:        %d"%mean_ages_math)
# how many different job categories
list_Fjob_catagories = list(map(lambda x:x['Fjob'],students_math))
list_Mjob_catagories = list(map(lambda x:x['Mjob'],students_math))
set_of_all_jobs = set(list_Fjob_catagories+list_Mjob_catagories)
print("Number of Different Job Categories: %d"%len(set_of_all_jobs))
print("Job Categories: ")
print(set_of_all_jobs)
Dalc_math = sorted(list(map(lambda x: int(x['Dalc']),students_math)))
Walc_math = sorted(list(map(lambda x: int(x['Walc']),students_math)))
math_Walc_mean = (ft.reduce(lambda a,b: a+b,Walc_math))/len(Walc_math)
math_Dalc_mean = (ft.reduce(lambda a,b: a+b,Dalc_math))/len(Dalc_math)

print("\nWeek Day alchol comsumtion average: %d"%math_Dalc_mean)
print("Lowest weekDay alcohol consumpsion: %d"%Dalc_math[0])
print("Highest weekDay alcohol consumpsion: %d"%Dalc_math[-1])
print("\nWeekend alchol comsumtion average:  %d"%math_Walc_mean)
print("Lowest weekend alcohol consumpsion: %d"%Walc_math[0])
print("Highest weekend alcohol consumpsion: %d"%Walc_math[-1])

absences_math = list(map(lambda x:int(x['absences']),students_math))
absences_math = sorted(absences_math)
print("\nabsences information")
print("Lowest Number of absences:  %d"%absences_math[0])
print("Highest Number of absences: %d"%absences_math[-1])
print("Range of absences:          %d"%(absences_math[-1]-absences_math[0]))


print('\nfinal Grade information')
f_grades_math = sorted(list(map(lambda x:int(x['G3']),students_math)))
print("lowest final grade:   %d"%f_grades_math[0])
print("highest final grade:  %d"%f_grades_math[-1])
print('range of final grades:%d'%(f_grades_math[-1]-f_grades_math[0]))


print('\nreasons for attending the school')
math_reasons_for_attendence = set(list(map(lambda x:x['reason'],students_math)))
print("different reasons for attendence count: %d"%len(math_reasons_for_attendence))
print('Reasons for attending school list: ')
print(math_reasons_for_attendence)
```

**Brief description of the data set:**

This data set (https://www.kaggle.com/uciml/student-alcohol-consumption) provides information about student alcohol consumption and several factors that could lead to an increase in drinking among the youth. The data was collected by a survey conducted by two university students in Italy who surveyed local secondary school students who took either math or Portuguese language classes. The dataset includes information about the students family situations, the students age and performance in the class and the amount of alcohol they typically consume on weekends and on weekdays on a scale from 1 to 5.  It is also worth noting that the data set is spit up into different tables one for each class meaning there is a table for the math students surveyed and a table for the Portuguese language students surveyed.

**Description of each attribute:**

**Variable Name	- Variable Description**

school - The school of which the student attends.  MS for Mousinho da Silveira and GP for Gabreil Pereira.

sex - The Students Gender Male(M) or Female(F). 

age - The Students age in years

address - The environment the student lives in Rural(R) and Urban(U)

famsize - The Size of the students family, GT3 if the students family is greater than 3 members

And LT3 if the student's family is less than 3 members.

Pstatus - The living status of the students' parents.  If they are living together then (T) and if the living apart(A)

Medu - The Mothers' education on a scale from 1 to 5 where 1 represents only completing up to primary education( grade 4) and 5 

represents completing some sort of higher education. 

Fedu - The Fathers' education on a scale from 1 to 5 where 1 represents only completing up to primary education( grade 4) and 5 

represents completing some sort of higher education. 

Mjob - The mothers’ job occupation(the sector she works in). Can be one of the following (Teacher, Health(works in healthcare), 

Service(works as civil servant),at_home, other)

Fjob - The Fathers’ job occupation(the sector he works in). Can be one of the following (Teacher, Health(works in healthcare), 

Service(works as civil servant),at_home, other)

reason - Reason for choosing to enroll in their school

guardian - Student’s current guardian (‘mother’, ‘father’ or ‘other’)

traveltime - Time it takes to get to school (1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)

studytime - Time spent studying each week (1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)

failures - Amount of past class failures 

schoolsup - Receiving extra educational support (yes/no)

paid - Taking extra paid classes within the courses’ subject (yes/no)

activities - Taking extra-curricular activities (yes/no)

nursery  - Attended nursery school (yes/no)

higher - Wants to take higher education (yes/no)

internet - Access to the internet at home (yes/no)

romantic - In a romantic relationship (yes/no)

famrel - Quality of family relationships (1 -worst to 5 -best)

freetime - Free time after school (1 -least to 5 -most)

goout - Amount of time spent out with friends (1 -least to 5 -most)

Dalc - Workday alcohol consumption (1 -least to 5 -most)

Walc - Weekend alcohol consumption (1 -least to 5 -most)

health - Current state of health (1 -worst to 5 -best)

absences - Number of absences from school (0-93)

G1 - First period grade (0-20)

G2 - Second period grade (0-20)

G3 - Final grade (0-20)
